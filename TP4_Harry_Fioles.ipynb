```python
import numpy as np
import random 
import copy

content= np.asanyarray([1,2,3,4])

# 1 : poison
# 2 : vin 
# 3 : avancer 
# 4 : reculer


def preparation(harry):
    while(harry[0]==2):
        random.shuffle(harry)
    
    i=-1
    while(i<len(harry)-2):
        i+=1
        if(harry[i]==2 and harry[i+1]==2):
            random.shuffle(harry)   
            i=-1
        if(harry[0]==2):
            random.shuffle(harry)   
            i=-1
    return harry

def preparation2(harry):
    for i in range(0,len(harry)):
        if(harry[2]!=2):
            for j in range(0,len(harry)):
                if(harry[j]==2 and j!=5):
                    permu=harry[2]
                    harry[2]=harry[j]
                    harry[j]=permu
        if(harry[5]!=2):
            for j in range(0,len(harry)):
                if(harry[j]==2 and j!=2):
                    permu=harry[5]
                    harry[5]=harry[j]
                    harry[j]=permu
    return harry
                

def condition2(harry):
    for i in range(0,len(harry)):
        if(harry[i]==2 and harry[i-1]!=1):
            for j in range(0,len(harry)-1):
                if(harry[j]==1 and harry[j+1]!=2):
                   permu=harry[i-1]
                   harry[i-1]=harry[j]
                   harry[j]=permu     
    return harry


def condition3(harry,cp):
    
    for i in range(0,len(harry)-1):
        if(harry[i]==1 and harry[i+1]==2):
            cp[i]="nan"
            cp[i+1]="nan"
            
    if(harry[0]==3 or harry[6]==3):
        if(cp[0]==3):
            for i in range(0,len(harry)):
                 if(cp[i]!="nan" and cp[i]!=3):
                     remp=harry[0]
                     harry[0]=harry[i]
                     harry[i]=remp
        if(cp[6]==3):
            for i in range(0,len(harry)):
                 if(cp[i]!="nan" and cp[i]!=3 and i!=0):
                     remp=harry[6]
                     harry[6]=harry[i]
                     harry[i]=remp
                     break
    return harry

def condition4(harry,cp):
    if(harry[0]==harry[6]):
        for i in range(0,len(harry)-1):
            if(harry[i]==1 and harry[i+1]==2):
                cp[i]="nan"
                cp[i+1]="nan"
    
        for i in range(0,len(harry)):
            if(cp[6]!="nan"):
                for j in range(0,len(harry)):
                    if(cp[j]!="nan" and cp[j]!=harry[6] and cp[j]!=3):
                        remp=harry[6]
                        harry[6]=harry[j]
                        harry[j]=remp
    return harry
```

```python
harry=[1,1,1,2,2,3,4]
random.shuffle(harry)

harry=preparation(harry)
harry=preparation2(harry)
harry=condition2(harry)

cp=copy.copy(harry)
harry=condition3(harry,cp)
harry=condition4(harry,cp)



            
for i in range(0,len(harry)):
    if(harry[i]==1):
        harry[i]='poison'
    if(harry[i]==2):
        harry[i]='vin'
    if(harry[i]==3):
        harry[i]='avancer'
    if(harry[i]==4):
        harry[i]='reculer'
            
            
            
print(harry)
```
